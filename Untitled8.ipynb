```python
import pandas as pd
import numpy as np
df = pd.read_csv('BMW_Car_Sales_Classification.csv')
df.head()
```

```python
df['Revenue'] = df['Price_USD'] * df['Sales_Volume']
```

```python
df.head()
```

```python
from datetime import datetime

current_year = datetime.now().year
df['Car_Age'] = current_year - df['Year']
```

```python
df.head()
```

```python
df['Price_Category'] = pd.cut(
    df['Price_USD'],
    bins=[0,30000,60000,100000,200000],
    labels=['Budget','Mid-Range','Premium','Luxury']
)
```

```python
df.head()
```

```python
df['Mileage_Category'] = pd.cut(
    df['Mileage_KM'],
    bins=[0,20000,50000,100000,200000],
    labels=['Low','Medium','High','Very High']
)
```

```python
df.head()
```

```python
total_revenue = df['Revenue'].sum()
print(total_revenue)
```

```python
total_sales = df['Sales_Volume'].sum()
print(total_sales)
```

```python
avg_price = df['Price_USD'].mean()
print(avg_price)
```
